```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(purchase_data)
purchase_df
```

## Player Count

* Display the total number of players


```python
unique_players = purchase_df["SN"].nunique()
unique_players
un_table = [{"Total Players": unique_players}]
unique_df = pd.DataFrame(data=un_table)
unique_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
all_items = len(purchase_df.index)

unique_items = purchase_df["Item ID"].nunique()

sum_price = purchase_df["Price"].sum()

average_price = sum_price/all_items

# (all_items, unique_items, sum_price, average_price)

summary = [{
    "Total Sales": sum_price,
    "Transactions": all_items,
    "Unique Items": unique_items,
    "Average Item Price": average_price
}]
summary_df = pd.DataFrame(data=summary)
summary_df
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
gender_group = purchase_df.groupby("Gender")
purchase_count = gender_group["Age"].count()
avg_purchase = gender_group["Price"].mean()
purchase_total = gender_group["Price"].sum()
unique_purchaser = gender_group["SN"].nunique()
avg_purchase_SN = ((gender_group["Price"].sum())/(unique_purchaser))

summary_gender_df = pd.DataFrame({"Purchase Count": purchase_count,
                      "Avg Price Purchase": avg_purchase,
                      "Total Purchases": purchase_total,
                      "Average Purchase by SN": avg_purchase_SN
    
})
summary_gender_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
labels = ["Under 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_df["Age Group"] = pd.cut(purchase_df["Age"], bins, labels=labels, include_lowest=True)

names_df = purchase_df.groupby(["Age Group"])
purchase_count = purchase_df["Age Group"].value_counts()
purchase_count

total_purchases = names_df["Price"].sum()
total_purchases

avg_purchases = (total_purchases/purchase_count)
avg_purchases

avg_purchase_pp = (total_purchases/un_names_df)
avg_purchase_pp

age_purchase_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                         "Average Purchase Price": avg_purchases,
                                         "Total Purchase Value": total_purchases,
                                         "Avg Total Purchase per Person": avg_purchase_pp
    })
age_purchase_analysis_df
# avg_purchase_price = 
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_group_df = purchase_df.groupby("SN")

SN_count = SN_group_df["SN"].value_counts()
SN_count

purchase_count = SN_group_df["SN"].count()
purchase_count

total_spent = SN_group_df["Price"].sum()
total_spent

SN_spent_df = pd.DataFrame(total_spent)
SN_spent_df
top_five_df = SN_spent_df.nlargest(5, "Price", keep="first")
top_five_df["Purchase Count"] = purchase_count
top_five_df["Avg Purchase Price"] = ((total_spent)/(purchase_count))
top_five_df["Total Purchase Value"] = total_spent 
top_five_df
```
